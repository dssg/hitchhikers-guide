```python
# Note: even though we are not using matplotlib directly
# we can still configure the settings through it since
# pandas uses matplotlib under the hood

# use this directive to embed plots in the notebook
%matplotlib inline

# import matplotlib
import matplotlib.pyplot as plt
import matplotlib

# use this to change the default style,
# use matplotlib.style.available to see
# available styles
matplotlib.style.use('ggplot')

# change the size of the plots
plt.rcParams['figure.figsize'] = (15, 5)

# import pandas
import pandas as pd
```

```python
# lets get some sample data
df = pd.read_csv('https://forge.scilab.org/index.php/p/rdataset/source/file/master/csv/datasets/iris.csv',
                 index_col=0)
# remove the dots in the columns and
#convert all to lowercase
df.columns = [col.replace('.', '_').lower() for col in df.columns]
```

```python
df.head()
```

```python
# plot to columns
df.sepal_length.plot()
df.sepal_width.plot()
```

```python
# histogram plot
df.petal_width.plot.hist()
```

```python
#histogram
hist = df.hist(column='petal_width', by='species', sharex=True, sharey=True)
```

```python
# box plot
df.plot.box()
```
