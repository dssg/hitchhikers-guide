```python
import folium
from folium import plugins
```

## Create a map

```python
# lat, long for Chicago
chicago = (41.875349,-87.6337)
# create a map, center in chicago and set a zoom level
m = folium.Map(location=chicago, zoom_start=12)
# you can move around!
m
```

## Adding markers

```python
millenium_park = (41.8825524,-87.6247401)
university_of_chicago = (41.7886079,-87.600902)

# add markers using lat long and add them
# to map m
folium.Marker(millenium_park).add_to(m)
folium.Marker(university_of_chicago).add_to(m)

# show map
m
```

## Adding color and style to markers

```python
# create a green icon with a check sign
icon = folium.Icon(color='green', icon='check')
# add the new marker
some_place = (41.9171045,-87.6370545)
folium.Marker(some_place, icon=icon).add_to(m)
# show the map
m
```

## Heatmap

```python
# this is just an utility function to
# generate random coordinates around a center
from random_coord import random_coord
```

```python
# generate 500 random points around chicago
coords = [random_coord(*chicago, distance=5000) for i in range(500)]
```

```python
# create a heatmap using the coords and add the layer to the map
heat_map = plugins.HeatMap(coords, min_opacity=0.1, radius=20, blur=5)
m.add_children(heat_map)
```
