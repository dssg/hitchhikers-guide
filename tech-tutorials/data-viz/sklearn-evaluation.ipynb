```python
# Note: even though we are not using matplotlib directly
# we can still configure the settings through it since
# sklearn-evaluation uses matplotlib under the hood
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
plt.rcParams['figure.figsize'] = (10, 10)

from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
from sklearn import datasets


from sklearn_evaluation import plot, table
```

```python
# Import some data
data = datasets.make_classification(1000, 10, 5, class_sep=0.7, n_classes=2)
X = data[0]
y = data[1]

# shuffle and split training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3,
                                                    random_state=0)
```

```python
# Train the model
est = RandomForestClassifier()
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
y_score = est.predict_proba(X_test)
y_true = y_test
```

```python
plot.confusion_matrix(y_true, y_pred, normalize=True)
```

```python
plot.feature_importances(est.feature_importances_)
```

```python
table.feature_importances(est.feature_importances_)
```

```python
plot.precision_recall(y_true, y_score)
```

```python
plot.roc(y_true, y_score)
```
